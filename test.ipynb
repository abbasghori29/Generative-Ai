```python
from scholarly import scholarly

def get_scholar_articles(query, num_articles=20):
    search_query = scholarly.search_pubs(query)
    articles = []
    for _ in range(num_articles):
        try:
            article = next(search_query)
            articles.append(article)
        except StopIteration:
            break
    return articles
query = "Ai in dental field"
articles = get_scholar_articles(query, num_articles=30)
```

```python
links=[]
for article in articles:
    if 'pub_url' in article:
        links.append(article['pub_url'])
    if 'eprint_url' in article:
        links.append(article['eprint_url'])
    if 'url' in article:
        links.append(article['url'])
links
```

```python
from langchain_community.document_loaders import WebBaseLoader
result=[]
for link in links:
    loader = WebBaseLoader(link)
    data = loader.load()
    result.append(data)
```
